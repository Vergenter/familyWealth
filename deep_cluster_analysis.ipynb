## Real wealth factor - cluster analysis

```python
from pandas import read_csv
from matplotlib import pyplot as plt
from pandas import DataFrame, MultiIndex

filename = 'dataset//wealth_factor.csv'
df_wealth_factor = read_csv(filename,index_col=["Country","Field"])
df_wealth_factor_total = df_wealth_factor[df_wealth_factor.index.map(lambda x: x[1]=="Total")].reset_index(level=1,drop=True)

filename = 'dataset//wealth_factor_total_labels.csv'
wealth_factor_total_labels = read_csv(filename,index_col="Country")
```

### Real wealth factor in cluters by year

```python
fig, axes = plt.subplots(wealth_factor_total_labels.columns.shape[0], figsize=(20,20),constrained_layout=True)
plt.suptitle("Real wealth factor in cluters by year",fontsize=20)
for column,ax in zip(wealth_factor_total_labels.columns, axes):
    ax.title.set_text(column)
    df_wealth_factor_total[wealth_factor_total_labels[column]].T.plot(ax=ax)
```

### Conclusions
Real wealth factor clustering looks reasonable, splitting countires into four groups with similar values.<br>
From all the countries the best result has **Germany**, that is only in the best prospering **group 1**.<br>
Countries in **group 2** have real wealth factor above 1, meaning that with typical monthly purchases it is possible to save money in them. Example countries in this group: **Austria, Canada, United Kingdom**.<br>
Countries in **group 3** have real wealth factor around 1, meaning that with modest monthly purchases it is possible to save money in them. Example countries in this group: **Poland, Slovakia, Greece**.<br>
Countries in **group 4** have real wealth factor below 1, meaning that even with modest monthly purchases it is hard to save money. Example countries in this group: **Ukraine, Panama, Philippines**.<br>
<br>
Looking at those graphs we can see a decrease of real wealth factor for Germany in 2014.<br>
In that year there is also a peak for Greece and Austria.
